## Task 3: Create Examples

```python
for class_id, values in emojis.items():
    png_file = Image.open(os.path.join('emojis', values['file'])).convert('RGBA')
    png_file.load()
    new_file = Image.new("RGB", png_file.size, (255, 255, 255))
    new_file.paste(png_file, mask=png_file.split()[3])
    emojis[class_id]['image'] = new_file
```

```python
emojis
```

```python
def create_example():
    class_id = np.random.randint(0,9) #from 0 - 8
    image = np.ones((144,144,3)) * 255 #create dummy image (white background with double-size of the emojis, 72x72)
    
    #pick a random point to place (horizontally - x axis) an emoji images on the white background randomly
    row = np.random.randint(0,72)
    
    #pick a random point to place (vertically - y axis) an emoji images on the white background randomly
    col = np.random.randint(0,72)
    
    #place an emoji image to random area
    image[row: row + 72, col: col+72] = np.array(emojis[class_id]['image'])
    
    #return image in int, class_id, normalized cropped x axis of the image ~10px, normalized cropped y axis of the image ~10px
    return image.astype('uint8'), class_id, (row + 10)/144, (col + 10)/144
```

```python
image, class_id, row, col = create_example()
plt.imshow(image)
```

## Task 4: Plot Bounding Boxes

```python
def plot_bounding_box(image, real_coords, pred_coords=[], normalized=False):
    if normalized:
        image *= 255.0
        image = image.astype('uint8')
        
    #convert array to image using PIL.Image.fromarray
    image = Image.fromarray(image)
    
    #prep to draw bounding boxes in the image
    draw = ImageDraw.Draw(image)
    
    #get coordinates of bounding boxes
    row, col = real_coords
    
    #scale back to 144
    row *= 144
    col *= 144
    
    #drawing real bounding boxes draw.rectangle((x1, y1, x2, y2))
    draw.rectangle((col, row, col+52, row+52), outline='green', width=3) #+52 because we'll crop the white space
    
    if len(pred_coords) == 2:
        pred_x, pred_y = pred_coords
        pred_x *= 144
        pred_y *= 144
        draw.rectangle((pred_x, pred_y, pred_x+52, pred_y+52), outline='blue', width=3)
    
    return image
```

```python
image = plot_bounding_box(image, real_coords=(row, col))
plt.imshow(image)
plt.title(emojis[class_id]['name'])
plt.show()
```

## Task 5: Data Generator

```python
def data_generator(batch_size=16):
    
    #prepare place holders
    while True:
        x_batch = np.zeros((batch_size, 144, 144, 3)) #train data
        y_batch = np.zeros((batch_size, 9)) #label data
        bbox_batch = np.zeros((batch_size, 2)) #2 for x, y coordinates
    
    #load the images to the place holders batch
        for i in range(0, batch_size):
            image, class_id, row, col = create_example()
            x_batch[i] = image / 255.0
            y_batch[i, class_id] = 1.0
            bbox_batch[i] = np.array([row, col])

        yield {'image': x_batch},{'class_out': y_batch, 'box_out': bbox_batch}
```

## Task 6: Model

```python
input_ = Input(shape=(144,144,3), name='image') #map the name from data generator to name= 
x = input_

#constructing 5 conv blocks
for i in range(0, 5):
    number_filters = 2**(4+i)
    x = Conv2D(number_filters, kernel_size = 3, activation = 'relu')(x)
    x = BatchNormalization()(x)
    x = MaxPool2D(2)(x) #2x2 pooling

x = Flatten()(x)
x = Dense(256, activation='relu')(x)

class_out = Dense(9, activation='softmax', name='class_out')(x) #map the name from data generator to name=
box_out = Dense(2, name='box_out')(x) #no need activation function for this

model = tf.keras.models.Model(input_, [class_out, box_out])
model.summary()
```

## Task 7: Custom Metric: IoU

```python
class IoU(tf.keras.metrics.Metric):
    def __init__(self, **kwargs):
        super(IoU, self).__init__(**kwargs)
        
        self.iou = self.add_weight(name='iou', initializer = 'zeros')
        self.total_iou = self.add_weight(name='total_iou', initializer = 'zeros')
        self.num_ex = self.add_weight(name='num_ex', initializer = 'zeros')
    
    def update_state(self, y_true, y_pred, sample_weight=None):
        def get_box(y):
            rows, cols = y[:, 0], y[:, 1]
            rows, cols = rows*144, cols*144 #scale up
            y1, y2 = rows, rows + 52
            x1, x2 = cols, cols + 52
            return x1,y1,x2,y2
        
        def get_area(x1, y1, x2, y2):
            return tf.math.abs(x1-x2) * tf.math.abs(y1-y2)
        
        real_x1, real_y1, real_x2, real_y2 = get_box(y_true)
        pred_x1, pred_y1, pred_x2, pred_y2 = get_box(y_pred)
        
        #get intersection box coordinates
        intersect_x1 = tf.maximum(real_x1, pred_x1) #for intersection x1, which one is greater?
        intersect_y1 = tf.maximum(real_y1, pred_y1) #for intersection y1, which one is greater?
        intersect_x2 = tf.minimum(real_x2, pred_x2) #for intersection x2, which one is smaller?
        intersect_y2 = tf.minimum(real_y2, pred_y2) #for intersection y2, which one is smaller?
        
        #get intersection area
        intersect_area = get_area(intersect_x1, intersect_y1, intersect_x2, intersect_y2)
        
        #get union area
        union_area = (get_area(real_x1, real_y1, real_x2, real_y2) + get_area(pred_x1, pred_y1, pred_x2, pred_y2)) - intersect_area
        
        #Calculate IOU
        iou = tf.math.divide(intersect_area,union_area)
        
        self.num_ex.assign_add(1) #add batch
        self.total_iou.assign_add(tf.reduce_mean(iou)) #average of the batch
        self.iou = tf.math.divide(self.total_iou, self.num_ex)
    
    def result(self):
        return self.iou
    def reset_state(self):
        self.iou = self.add_weight(name='iou', initializer = 'zeros')
        self.total_iou = self.add_weight(name='total_iou', initializer = 'zeros')
        self.num_ex = self.add_weight(name='num_ex', initializer = 'zeros')
```

## Task 8: Compile the Model

```python
model.compile(
    loss={
        'class_out': 'categorical_crossentropy',
        'box_out': 'mse'
    },
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    metrics={
        'class_out': 'accuracy',
        'box_out': IoU(name='iou')
    }
)
```

## Task 9: Custom Callback: Model Testing

```python
def test_model(model, test_datagen):
    example, label = next(test_datagen)
    x = example['image']
    y = label['class_out'] #get the actual class_id
    box = label['box_out']
    
    pred_y, pred_box = model.predict(x)
    pred_coords = pred_box[0]
    real_coords = box[0]
    pred_class = np.argmax(pred_y[0])
    image = x[0]
    
    real_class_name = emojis[np.argmax(y[0])]['name']
    pred_class_name = emojis[pred_class]['name']
    
    image = plot_bounding_box(image, real_coords=real_coords, pred_coords=pred_coords, normalized=True)
    color = 'green' if real_class_name == pred_class_name else 'red'
    
    plt.imshow(image)
    plt.xlabel(f'Pred: {pred_class_name}', color=color)
    plt.ylabel(f'Truth: {real_class_name}', color=color)
    plt.xticks([])
    plt.yticks([])
```

```python
def test(model):
    test_datagen = data_generator(1)
    
    plt.figure(figsize=(16,4))
    for i in range(0,6):
        plt.subplot(1,6, i+1)
        test_model(model, test_datagen)
    plt.show()
```

```python
test(model)
```

```python
class ShowTestImages(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        test(self.model)
```

## Task 10: Model Training

```python
def lr_schedule(epoch, lr):
    if (epoch + 1) % 5 == 0:
        lr *= 0.2
    return max(lr, 3e-7)

_ = model.fit(
    data_generator(),
    epochs=50,
    steps_per_epoch=500,
    callbacks=[
        ShowTestImages(),
        tf.keras.callbacks.EarlyStopping(monitor='box_out_iou', patience=3, mode='max'), #stop if iou didnt increase for 4 consecutive epochs
        tf.keras.callbacks.LearningRateScheduler(lr_schedule)
    ]
)
```
